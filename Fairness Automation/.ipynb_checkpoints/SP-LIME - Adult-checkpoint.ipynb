```python
# Import necessary libraries
import lime
import lime.lime_tabular
from pmlb import fetch_data
import sklearn.ensemble
import sklearn.linear_model
import sklearn.model_selection
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import r2_score
import pandas as pd

# Set random seed for reproducibility
np.random.seed(1)

# Load the Adult dataset
adult = fetch_data('adult')
X = adult.drop(columns=['target', 'fnlwgt', 'native-country'])  # Drop unnecessary columns
y = adult['target']  # Target variable: whether income >50K (encoded as 0 or 1)

# Convert the target to float for regression
y = y.astype(float)

# Preprocess categorical features by encoding them as numeric
categorical_cols = X.select_dtypes(include=['object']).columns
for col in categorical_cols:
    X[col] = LabelEncoder().fit_transform(X[col])

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame to retain column names
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# Split the data into train and test sets
train, test, labels_train, labels_test = sklearn.model_selection.train_test_split(X_scaled, y, train_size=0.80, test_size=0.20)

# Train a RandomForestRegressor
rf = sklearn.ensemble.RandomForestRegressor(n_estimators=1000)
rf.fit(train, labels_train)

# Train a LinearRegression model
lr = sklearn.linear_model.LinearRegression()
lr.fit(train, labels_train)

# Calculate and print the R^2 score of both models on the test set
rf_r2 = r2_score(labels_test, rf.predict(test))
lr_r2 = r2_score(labels_test, lr.predict(test))

print(f"Random Forest R^2 Score: {round(rf_r2, 3)}")
print(f"Linear Regression R^2 Score: {round(lr_r2, 3)}")

# ----- LIME EXPLAINER -----

# Identify categorical features by their column indices in the processed dataframe (X_scaled)
categorical_features = [X.columns.get_loc(col) for col in categorical_cols]

# Create an explainer for LIME
explainer = lime.lime_tabular.LimeTabularExplainer(
    train.values,  # Ensure we pass the data as a NumPy array
    feature_names=X.columns,  # Feature names from the dataset
    class_names=['income'],  # Class name (for this task, we treat it as "income")
    categorical_features=categorical_features,  # Categorical feature indices
    verbose=True,
    mode='regression',
    discretize_continuous=False  # Don't discretize continuous variables
)

# Pick a sample from the test set to explain
i = 5  # Index of the test instance to explain
exp = explainer.explain_instance(test.values[i], rf.predict, num_features=10)

# Display the explanation
exp.show_in_notebook(show_table=True)
```

```python
import warnings
from lime import submodular_pick

# Apply Submodular Pick to generate multiple diverse explanations
sp_obj = submodular_pick.SubmodularPick(
    explainer, 
    train.values,  # Pass training data as a NumPy array
    rf.predict,  # Model's prediction function
    sample_size=20,  # Number of instances to sample for explanations
    num_features=14,  # Number of features to explain per instance
    num_exps_desired=5  # Number of explanations to select
)

# Visualize the explanations generated by Submodular Pick
for exp in sp_obj.sp_explanations:
    exp.as_pyplot_figure()
```

```python
import pandas as pd

# Convert explanations into a DataFrame
W = pd.DataFrame([dict(exp.as_list()) for exp in sp_obj.sp_explanations])

# Display the DataFrame
print(W.head())
```

```python
import matplotlib.pyplot as plt

# Calculate the mean importance for each feature across all explanations
feature_importance_mean = W.mean()

# Sort the features by importance
feature_importance_mean = feature_importance_mean.sort_values(ascending=False)

# Plot the feature importance as a bar plot
plt.figure(figsize=(10, 6))
feature_importance_mean.plot(kind='bar', color='skyblue')

# Add titles and labels
plt.title("Average Feature Importance Across LIME Explanations")
plt.xlabel("Features")
plt.ylabel("Average Importance")

# Display the plot
plt.tight_layout()
plt.show()
```
